```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import folium
```

```python
df = pd.read_csv('store-lat-long.csv')
df.head()
```

```python
map = folium.Map(location=[-35.334308, 149.171660], zoom_start=10, control_scale=True)
```

```python
for index, location_info in df.iterrows():
    folium.Marker([location_info["StoreLatitude"], location_info["StoreLongitude"]], popup=location_info["StoreName"]).add_to(map)
```
